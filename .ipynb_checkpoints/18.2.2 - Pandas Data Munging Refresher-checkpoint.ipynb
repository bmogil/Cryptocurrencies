```python
import pandas as pd
```

```python
iris_df = pd.read_csv("Resources/iris.csv")
iris_df.head(10)
```

```python
new_iris_df = iris_df.drop(["class"], axis=1)
new_iris_df.head(10)
```

```python
column_names = ["petal_length", "sepal_length", "petal_width", "sepal_width"]

new_iris_df = new_iris_df.reindex(columns=column_names)

new_iris_df.head(10)
```

```python
output_file_path = "Resources/new_iris_data.csv"

new_iris_df.to_csv(output_file_path, index=False)
```
